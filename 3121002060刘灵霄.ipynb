```python
import numpy as np
from scipy.optimize import minimize


def objective_function(x):
    return x[0]**2 + x[1]**2 + 3*x[0] + 4*x[1] - 26


def gradient_function(x):
    return np.array([2*x[0] + 3, 2*x[1] + 4])


initial_values = np.array([3, 2])


result_ncg = minimize(objective_function, initial_values, jac=gradient_function)

print("牛顿共轭梯度法的最优解：", result_ncg.x)
print("最小值：", result_ncg.fun)
```

```python
result_bfgs = minimize(objective_function, initial_values,jac=gradient_function)

print("拟牛顿法（BFGS）的最优解：", result_bfgs.x)
print("最小值：", result_bfgs.fun)
```

```python
from scipy import interpolate
import matplotlib.pyplot as plt


x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 1, 4, 9, 16])


power_function = interpolate.interp1d(x, y, kind='quadratic')


chebyshev_function = interpolate.CubicSpline(x, y)


x_dense = np.linspace(0, 4, 100)
y_power = power_function(x_dense)
y_chebyshev = chebyshev_function(x_dense)


plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x_dense, y_power, label='Power Function Interpolation')
plt.plot(x_dense, y_chebyshev, label='Chebyshev Function Interpolation')
plt.title('Interpolation with Power and Chebyshev Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.sparse import csr_matrix


data = np.array([1, 0, 1, 1, 1])  
row_indices = np.array([0, 0, 1, 2, 1])  
col_indices = np.array([1, 0, 2, 1, 0]) 


B = csr_matrix((data, (row_indices, col_indices)), shape=(3, 3))

print(B.toarray())
```

```python
import pandas as pd


data = {'number': [34, 21, 29, 56],
        'color': ['red', 'brown', 'yellow', 'red']}
index = ['apple', 'dates', 'banana', 'cherry']
df = pd.DataFrame(data, index=index)


scores = {'apple': 5, 'dates': 3, 'banana': 3, 'cherry': 2}
df['score'] = df.index.map(scores)


df = df.sort_values(by='number', ascending=False)

print(df)
```
